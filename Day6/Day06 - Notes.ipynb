## 6.1 Introduction to Polymorphism

**Polymorphism** (Greek: *poly* = many, *morph* = form) allows the same function name or operator to behave differently based on context.

### Two Types of Polymorphism

```c++
Polymorphism
              /            \
     Compile-time           Runtime
   (Static Binding)      (Dynamic Binding)
         |                      |
    - Function Overloading    - Virtual Functions
    - Operator Overloading    - Method Overriding
```

| Feature | Compile-Time | Runtime |
|---|---|---|
| **When resolved** | At compile time | At runtime |
| **Speed** | Faster (no overhead) | Slightly slower (vtable lookup) |
| **Flexibility** | Less flexible | More flexible |
| **Implementation** | Overloading, templates | Virtual functions, overriding |
| **Keyword** | — | `virtual` |

## 6.2 Compile-Time Polymorphism (Static Binding)

### 6.2.1 Function Overloading

Multiple functions with the **same name** but **different parameters** (type, count, or order).

#### Rules of Function Overloading

```c++
// Valid overloads — differ in number/types of parameters
void print(int i);
void print(double d);
void print(string s);
void print(int i, string s);

// INVALID — only return type differs (compiler error)
int getValue();
double getValue();   // ERROR: ambiguous
```

#### Complete Example

```c++
#include <iostream>
#include <string>
using namespace std;

class Calculator {
    public:
        // Overloaded add functions
        int add(int a, int b) {
            cout << "int + int = ";
            return a + b;
        }
        
        double add(double a, double b) {
            cout << "double + double = ";
            return a + b;
        }
        
        int add(int a, int b, int c) {
            cout << "int + int + int = ";
            return a + b + c;
        }
        
        double add(int a, double b) {
            cout << "int + double = ";
            return a + b;
        }
        
        string add(string a, string b) {
            cout << "string + string = ";
            return a + b;
        }
};

int main() {
    Calculator calc;
    
    cout << calc.add(5, 10) << endl;          // int + int
    cout << calc.add(3.5, 2.7) << endl;       // double + double
    cout << calc.add(1, 2, 3) << endl;        // three ints
    cout << calc.add(5, 3.14) << endl;        // int + double
    cout << calc.add("Hello, ", "World!") << endl;  // string + string
    
    return 0;
}
```

**Output:**

```c++
int + int = 15
double + double = 6.2
int + int + int = 6
int + double = 8.14
string + string = Hello, World!
```

### Function Overloading vs Default Arguments

```c++
#include <iostream>
using namespace std;

class Demo {
    public:
        // Approach 1: Overloading
        void show(int a) { cout << "One int: " << a << endl; }
        void show(int a, int b) { cout << "Two ints: " << a << ", " << b << endl; }
        
        // Approach 2: Default arguments (simpler when applicable)
        void display(int a, int b = 0) {
            cout << "Display: " << a << ", " << (b ? to_string(b) : "default") << endl;
        }
};

int main() {
    Demo d;
    d.show(10);
    d.show(10, 20);
    d.display(10);
    d.display(10, 20);
    return 0;
}
```

**Output:**

```c++
One int: 10
Two ints: 10, 20
Display: 10, default
Display: 10, 20
```

### 6.2.2 Operator Overloading

**Operator overloading** allows C++ operators (like `+`, `-`, `<<`, `[]`) to work with user-defined types.

#### Syntax

```c++
returnType operator symbol (parameters) {
    // implementation
}
```

#### Operators That Can Be Overloaded

| Category | Operators |
|---|---|
| **Arithmetic** | `+` `-` `*` `/` `%` |
| **Relational** | `==` `!=` `<` `>` `<=` `>=` |
| **Assignment** | `=` `+=` `-=` `*=` `/=` |
| **I/O Stream** | `<<` `>>` |
| **Index** | `[]` |
| **Function call** | `()` |
| **Increment/Decrement** | `++` `--` |

#### Operators That CANNOT Be Overloaded

| Operator | Reason |
|---|---|
| `::` (scope resolution) | Operates on types, not values |
| `.` (member access) | Would break fundamental semantics |
| `.*` (pointer-to-member) | Same as above |
| `?:` (ternary) | Not an actual function call |
| `sizeof` | Built-in compile-time operator |
| `#` `##` (preprocessor) | Handled before compilation |

#### Complete Operator Overloading Example

```c++
#include <iostream>
using namespace std;

class Complex {
    private:
        double real;
        double imag;
    
    public:
        Complex(double r = 0, double i = 0) : real(r), imag(i) {}
        
        // 1. Arithmetic operator (+)
        Complex operator+(const Complex& other) const {
            return Complex(real + other.real, imag + other.imag);
        }
        
        // 2. Subtraction operator (-)
        Complex operator-(const Complex& other) const {
            return Complex(real - other.real, imag - other.imag);
        }
        
        // 3. Unary minus (negate)
        Complex operator-() const {
            return Complex(-real, -imag);
        }
        
        // 4. Equality operator (==)
        bool operator==(const Complex& other) const {
            return (real == other.real) && (imag == other.imag);
        }
        
        // 5. Prefix increment (++c)
        Complex& operator++() {
            ++real;
            return *this;
        }
        
        // 6. Postfix increment (c++) — dummy int parameter
        Complex operator++(int) {
            Complex temp = *this;
            ++real;
            return temp;
        }
        
        // 7. Subscript operator ([])
        double operator[](int index) const {
            if (index == 0) return real;
            if (index == 1) return imag;
            throw out_of_range("Index must be 0 (real) or 1 (imag)");
        }
        
        // 8. Friend function for output (<<)
        friend ostream& operator<<(ostream& os, const Complex& c);
        
        // 9. Friend function for input (>>)
        friend istream& operator>>(istream& is, Complex& c);
};

// Friend function definitions (outside class)
ostream& operator<<(ostream& os, const Complex& c) {
    os << c.real;
    if (c.imag >= 0)
        os << " + " << c.imag << "i";
    else
        os << " - " << (-c.imag) << "i";
    return os;
}

istream& operator>>(istream& is, Complex& c) {
    cout << "Enter real part: ";
    is >> c.real;
    cout << "Enter imag part: ";
    is >> c.imag;
    return is;
}

int main() {
    Complex c1(3, 4);
    Complex c2(1, 2);
    
    cout << "c1 = " << c1 << endl;
    cout << "c2 = " << c2 << endl;
    
    cout << "c1 + c2 = " << (c1 + c2) << endl;
    cout << "c1 - c2 = " << (c1 - c2) << endl;
    cout << "-c1 = " << (-c1) << endl;
    cout << "c1 == c2: " << (c1 == c2 ? "true" : "false") << endl;
    
    cout << "c1++ = " << (c1++) << ", after: " << c1 << endl;
    cout << "++c1 = " << (++c1) << endl;
    
    cout << "c1[0] = " << c1[0] << ", c1[1] = " << c1[1] << endl;
    
    return 0;
}
```

**Output:**

```c++
c1 = 3 + 4i
c2 = 1 + 2i
c1 + c2 = 4 + 6i
c1 - c2 = 2 + 2i
-c1 = -3 - 4i
c1 == c2: false
c1++ = 3 + 4i, after: 4 + 4i
++c1 = 5 + 4i
c1[0] = 5, c1[1] = 4
```

## 6.3 Runtime Polymorphism (Dynamic Binding)

### 6.3.1 Virtual Functions

A **virtual function** is a member function declared with the `virtual` keyword in the base class. When called via a base class pointer/reference, the **actual derived class version** is executed at runtime.

```c++
#include <iostream>
#include <string>
using namespace std;

class Animal {
    public:
        virtual void speak() {     // Virtual function
            cout << "Animal makes a sound" << endl;
        }
        
        void eat() {               // Non-virtual — no polymorphism
            cout << "Animal eats" << endl;
        }
        
        virtual ~Animal() {}       // Virtual destructor (important!)
};

class Dog : public Animal {
    public:
        void speak() override {    // Override virtual function
            cout << "Dog barks: Woof!" << endl;
        }
        
        void eat() {               // Hides base class eat(), not override
            cout << "Dog eats kibble" << endl;
        }
};

class Cat : public Animal {
    public:
        void speak() override {
            cout << "Cat meows: Meow!" << endl;
        }
};

int main() {
    Animal* animals[3];
    animals[0] = new Animal();
    animals[1] = new Dog();
    animals[2] = new Cat();
    
    cout << "=== Virtual functions (polymorphic) ===" << endl;
    for (int i = 0; i < 3; i++) {
        animals[i]->speak();  // Calls correct version based on object type
    }
    
    cout << "\n=== Non-virtual functions (NOT polymorphic) ===" << endl;
    for (int i = 0; i < 3; i++) {
        animals[i]->eat();    // ALWAYS calls Animal::eat()
    }
    
    // Cleanup
    for (int i = 0; i < 3; i++) {
        delete animals[i];
    }
    
    return 0;
}
```

**Output:**

```c++
=== Virtual functions (polymorphic) ===
Animal makes a sound
Dog barks: Woof!
Cat meows: Meow!

=== Non-virtual functions (NOT polymorphic) ===
Animal eats
Animal eats
Animal eats
```

### How Virtual Functions Work — The VTable

```c++
Object of class Dog:
+-----------------------+
| vtable pointer (vptr) | ----> Points to Dog's vtable
+-----------------------+
| Dog-specific data     |
+-----------------------+

Dog's vtable:
+-----------------------+
| Animal::eat()         |  (non-virtual, not in vtable)
+-----------------------+
| Dog::speak()          |  (overrides Animal::speak)
+-----------------------+
| Animal::~Animal()     |  (or Dog::~Dog() if overridden)
+-----------------------+
```

> **Key insight:** Each object has a hidden `vptr` (virtual pointer) that points to the class's **vtable** (virtual table). The vtable is an array of function pointers. Calling a virtual function at runtime looks up the vtable to find the correct function.

### 6.3.2 Pure Virtual Functions and Abstract Classes

A **pure virtual function** is a virtual function declared with `= 0`. A class containing at least one pure virtual function becomes **abstract** and **cannot be instantiated**.

```c++
#include <iostream>
#include <string>
using namespace std;

// Abstract class (can't be instantiated)
class Shape {
    protected:
        string color;
    
    public:
        Shape(string c) : color(c) {}
        
        // Pure virtual functions — must be overridden
        virtual double area() const = 0;
        virtual void draw() const = 0;
        
        // Concrete method
        string getColor() const { return color; }
        
        // Virtual destructor (always needed in base classes)
        virtual ~Shape() {}
};

class Circle : public Shape {
    private:
        double radius;
    
    public:
        Circle(string c, double r) : Shape(c), radius(r) {}
        
        double area() const override {
            return 3.14159 * radius * radius;
        }
        
        void draw() const override {
            cout << "○ Drawing a " << color << " circle (r=" << radius 
                 << ", area=" << area() << ")" << endl;
        }
};

class Square : public Shape {
    private:
        double side;
    
    public:
        Square(string c, double s) : Shape(c), side(s) {}
        
        double area() const override {
            return side * side;
        }
        
        void draw() const override {
            cout << "□ Drawing a " << color << " square (side=" << side 
                 << ", area=" << area() << ")" << endl;
        }
};

int main() {
    // Shape s("red");  // ERROR! Cannot instantiate abstract class
    
    Shape* shapes[2];
    shapes[0] = new Circle("Red", 5);
    shapes[1] = new Square("Blue", 4);
    
    for (int i = 0; i < 2; i++) {
        shapes[i]->draw();
    }
    
    for (int i = 0; i < 2; i++) {
        delete shapes[i];
    }
    
    return 0;
}
```

**Output:**

```c++
○ Drawing a Red circle (r=5, area=78.5397)
□ Drawing a Blue square (side=4, area=16)
```

## 6.4 `override` and `final` Keywords (C++11)

### `override`

The `override` specifier ensures that a virtual function is actually **overriding** a base class virtual function. If not, the compiler generates an error.

```c++
class Base {
    public:
        virtual void func(int x) { }
        virtual void display() { }
};

class Derived : public Base {
    public:
        // void func(int x) override { }       // ✅ OK: overrides Base::func
        void func(double x) { }                // ❌ BUG! Creates NEW function (not override)
        void func(int x) override { }          // ✅ Compiler checks and confirms override
        // void show() override { }             // ❌ ERROR: Base has no virtual show()
};
```

### `final`

The `final` specifier prevents further overriding (on functions) or further inheritance (on classes).

```c++
class Vehicle {
    public:
        virtual void start() { cout << "Vehicle start" << endl; }
};

class Car final : public Vehicle {   // 'final' class — cannot be inherited
    public:
        void start() override final {  // 'final' function — cannot be overridden further
            cout << "Car start" << endl;
        }
};

// class SportsCar : public Car { };   // ❌ ERROR: Car is final!
```

### Why Use `override` and `final`?

| Benefit | Explanation |
|---|---|
| **Catches errors early** | Misspelled function names → compiler error instead of silent wrong behavior |
| **Self-documenting** | Clearly expresses intent to override or prevent override |
| **Safer refactoring** | If base class changes, `override` will notify of broken code |
| **Optimization** | `final` can enable devirtualization optimizations |

## Coding Examples

### Example 1 — Polymorphic Payment System

```c++
#include <iostream>
#include <string>
#include <vector>
using namespace std;

// Abstract base class
class Payment {
    protected:
        double amount;
        string transactionId;
    
    public:
        Payment(double amt, string id) : amount(amt), transactionId(id) {}
        
        virtual void process() = 0;         // Pure virtual
        virtual string getType() const = 0; // Pure virtual
        
        virtual ~Payment() {}
};

class CreditCardPayment : public Payment {
    private:
        string cardNumber;
    
    public:
        CreditCardPayment(double amt, string id, string card) 
            : Payment(amt, id), cardNumber(card) {}
        
        void process() override {
            cout << "Processing Credit Card payment of $" << amount << endl;
            cout << "Card ending in: " << cardNumber.substr(cardNumber.length() - 4) << endl;
            cout << "Transaction: " << transactionId << " — APPROVED" << endl;
        }
        
        string getType() const override {
            return "Credit Card";
        }
};

class PayPalPayment : public Payment {
    private:
        string email;
    
    public:
        PayPalPayment(double amt, string id, string e) 
            : Payment(amt, id), email(e) {}
        
        void process() override {
            cout << "Processing PayPal payment of $" << amount << endl;
            cout << "PayPal account: " << email << endl;
            cout << "Transaction: " << transactionId << " — APPROVED" << endl;
        }
        
        string getType() const override {
            return "PayPal";
        }
};

class CryptoPayment : public Payment {
    private:
        string walletAddress;
    
    public:
        CryptoPayment(double amt, string id, string wallet) 
            : Payment(amt, id), walletAddress(wallet) {}
        
        void process() override {
            cout << "Processing Crypto payment of $" << amount << endl;
            cout << "Wallet: " << walletAddress.substr(0, 10) << "..." << endl;
            cout << "Transaction: " << transactionId << " — CONFIRMED" << endl;
        }
        
        string getType() const override {
            return "Cryptocurrency";
        }
};

int main() {
    vector<Payment*> payments;
    
    payments.push_back(new CreditCardPayment(150.00, "TXN-001", "1234567890123456"));
    payments.push_back(new PayPalPayment(89.99, "TXN-002", "arjun@email.com"));
    payments.push_back(new CryptoPayment(500.00, "TXN-003", "0x7421Fc9aB0d1E7b8A3fC6d9E2b4A8c0F1d3E5b7"));
    
    for (Payment* p : payments) {
        cout << "\n=== " << p->getType() << " ===" << endl;
        p->process();
    }
    
    for (Payment* p : payments) {
        delete p;
    }
    
    return 0;
}
```

**Output:**

```c++
=== Credit Card ===
Processing Credit Card payment of $150
Card ending in: 3456
Transaction: TXN-001 — APPROVED

=== PayPal ===
Processing PayPal payment of $89.99
PayPal account: arjun@email.com
Transaction: TXN-002 — APPROVED

=== Cryptocurrency ===
Processing Crypto payment of $500
Wallet: 0x7421Fc9aB0...
Transaction: TXN-003 — CONFIRMED
```

### Example 2 — Comparison: Virtual vs Non-Virtual

```c++
#include <iostream>
using namespace std;

class Base {
    public:
        void nonVirtual() { cout << "Base::nonVirtual" << endl; }
        virtual void virtualFunc() { cout << "Base::virtualFunc" << endl; }
        virtual ~Base() {}
};

class Derived : public Base {
    public:
        void nonVirtual() { cout << "Derived::nonVirtual" << endl; }
        void virtualFunc() override { cout << "Derived::virtualFunc" << endl; }
};

int main() {
    Base* ptr = new Derived();
    
    cout << "Called via Base* pointing to Derived object:" << endl;
    cout << "Non-virtual: "; ptr->nonVirtual();   // Base::nonVirtual (static binding)
    cout << "Virtual    : "; ptr->virtualFunc();   // Derived::virtualFunc (dynamic binding)
    
    cout << "\nCalled via reference:" << endl;
    Derived d;
    Base& ref = d;
    ref.nonVirtual();     // Base::nonVirtual (static binding)
    ref.virtualFunc();    // Derived::virtualFunc (dynamic binding)
    
    delete ptr;
    return 0;
}
```

**Output:**

```c++
Called via Base* pointing to Derived object:
Non-virtual: Base::nonVirtual
Virtual    : Derived::virtualFunc

Called via reference:
Base::nonVirtual
Derived::virtualFunc
```

### Example 3 — Virtual Destructor (Why It Matters)

```c++
#include <iostream>
using namespace std;

class BaseWithoutVirtual {
    public:
        ~BaseWithoutVirtual() { cout << "Base WITHOUT virtual destructor" << endl; }
};

class BaseWithVirtual {
    public:
        virtual ~BaseWithVirtual() { cout << "Base WITH virtual destructor" << endl; }
};

class DerivedA : public BaseWithoutVirtual {
    public:
        ~DerivedA() { cout << "DerivedA destructor" << endl; }
};

class DerivedB : public BaseWithVirtual {
    public:
        ~DerivedB() { cout << "DerivedB destructor" << endl; }
};

int main() {
    cout << "=== Without virtual destructor (BUG!) ===" << endl;
    BaseWithoutVirtual* badPtr = new DerivedA();
    delete badPtr;  // Only calls Base destructor! Memory leak!
    
    cout << "\n=== With virtual destructor (CORRECT) ===" << endl;
    BaseWithVirtual* goodPtr = new DerivedB();
    delete goodPtr;  // Calls DerivedB destructor, then Base destructor
    
    return 0;
}
```

**Output:**

```c++
=== Without virtual destructor (BUG!) ===
Base WITHOUT virtual destructor
               ^--- DerivedA destructor NOT called! MEMORY LEAK!

=== With virtual destructor (CORRECT) ===
DerivedB destructor
Base WITH virtual destructor
               ^--- Both destructors called! Proper cleanup!
```

## Common Mistakes

| Mistake | Wrong | Correct |
|---|---|---|
| **Forgetting `virtual` for overriding** | `void func()` in base; `void func()` in derived — only hides, doesn't override | Add `virtual` in base class declaration |
| **Missing `override` keyword** | `void func(int x) const` doesn't match `virtual void func(int x)` | Use `override` to catch signature mismatches |
| **Non-virtual destructor in base** | `~Base() {}` | `virtual ~Base() {}` |
| **Trying to instantiate abstract class** | `Shape s;` where Shape has pure virtual | Only instantiate concrete derived classes |
| **Confusing overloading with overriding** | Same name, different params ≠ override | Override requires **exact same signature** |
| **Calling virtual function in constructor/destructor** | Calls base version, not derived | Avoid virtual calls during construction/destruction |

## Interview Questions

1. **What is polymorphism and what are its types?**
   - Polymorphism: same interface, different behaviors. **Compile-time** (overloading, templates) and **Runtime** (virtual functions, overriding).

2. **What is the difference between function overloading and function overriding?**
   - **Overloading**: Same name, different parameters, compile-time, same class. **Overriding**: Same signature, runtime, base→derived class.

3. **What is a virtual function? How does it work internally?**
   - A function declared with `virtual` that can be overridden. Internally uses **vtable** (virtual table) — each object has a `vptr` that points to its class's vtable, which maps virtual functions to the correct implementation.

4. **What is a pure virtual function?**
   - A virtual function declared with `= 0`. Makes the class abstract (cannot be instantiated). Derived classes **must** override it.

5. **Why should base class destructors be virtual?**
   - So that `delete basePtr` where `basePtr` points to a derived object properly calls the derived destructor first, then the base destructor — preventing memory leaks.

## Quick Revision Summary

- **Compile-time polymorphism**: Function overloading, operator overloading (static binding)
- **Runtime polymorphism**: Virtual functions, method overriding (dynamic binding)
- **`virtual` keyword**: Enables runtime polymorphism
- **`override`**: Ensures a function actually overrides a base virtual function (C++11)
- **`final`**: Prevents further overriding or inheritance (C++11)
- **Pure virtual**: `virtual void func() = 0;` — makes class abstract
- **Abstract class**: Has ≥1 pure virtual function; cannot be instantiated
- **VTable**: Internal mechanism storing function pointers for virtual dispatch
- **Virtual destructor**: Always declare `virtual` in base classes
